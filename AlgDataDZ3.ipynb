Измените функцию calc_logloss так, чтобы нули по возможности не попадали в np.log.

```python
X = np.array([ [   1,    1,  500,    1],
               [   1,    1,  700,    1],
               [   1,    2,  750,    2],
               [   1,    5,  600,    1],
               [   1,    3, 1450,    2],
               [   1,    0,  800,    1],
               [   1,    5, 1500,    3],
               [   1,   10, 2000,    3],
               [   1,    1,  450,    1],
               [   1,    2, 1000,    2]], dtype = np.float64)

y = np.array([0, 0, 1, 0, 1, 0, 1, 0, 1, 1], dtype = np.float64)
```

```python
def standard_scale(x):
    res = (x - x.mean()) / x.std()
    return res
```

```python
X_st = X.copy()
X_st[:, 2] = standard_scale(X[:, 2])
```

```python
def calc_logloss(y, y_pred):
    y_pred = np.where(y_pred == 1, 0.9999999999, y_pred)
    y_pred = np.where(y_pred == 0, 0.0000000001, y_pred)
    err = - np.mean(y * np.log(y_pred) + (1.0 - y) * np.log(1.0 - y_pred))
    return err
```

```python
y1 = np.array([1, 0])
y_pred1 = np.array([1, 0.1])
calc_logloss(y1, y_pred1)
```

Подберите аргументы функции eval_model для логистической регрессии таким образом, чтобы log loss был минимальным.

```python
def sigmoid(z):
    res = 1 / (1 + np.exp(-z))
    return res
```

```python
def eval_model(X, y, iterations, eta=1e-4):
    np.random.seed(42)
    W = np.random.randn(X.shape[1])
    n = X.shape[0]
    
    for i in range(iterations):
        z = np.dot(X, W)
        y_pred = sigmoid(z)
        err = calc_logloss(y, y_pred)
        
        dQ = 1/n * X.T @ (y_pred - y)
        W -= eta * dQ
        if i % (iterations / 10) == 0:
            print(i, W, err)
    return W, err
```

```python
W, err = eval_model(X_st, y, iterations=100000, eta=1e-4)
```

```python
from tqdm import tqdm
iterations_ = np.arange(100, 10000)
err_ = 10000
err_list = []
# for iterations_ in np.arange(100, 10000):
for iterations_ in tqdm(iterations_):
    for eta_ in np.arange(1e-4, 1e-3):
        W, err = eval_model(X_st, y, iterations=iterations_, eta=eta_)
        err_list.append(err)
        if err < err_:
            err_ = err
            iterations1 = iterations_
            eta1 = eta_
            W_ = W
print(f'W={W_}, err={err_}, iterations={iterations1}, eta={eta1}')
```

```python
plt.title('Log loss')
plt.xlabel('time')
plt.ylabel('loss')
plt.plot(err_list);
```

Создайте функцию calc_pred_proba, возвращающую предсказанную вероятность класса 1

```python
def calc_pred_proba(W, X):
    y_pred_proba = sigmoid(np.dot(X, W))
    return y_pred_proba
```

```python
calc_pred_proba(W, X_st)
```

```python
y_pred_proba = calc_pred_proba(W, X)
y_pred_proba
```

Создайте функцию calc_pred, возвращающую предсказанный класс

```python
def calc_pred(W, X):
    y_pred_proba = calc_pred_proba(W, X)
    y_pred = np.where(y_pred_proba > 0.5, 1, 0)
    return y_pred
```

```python
y_pred = calc_pred(W, X_st)
y_pred
```

```python
def accuracy(y, y_pred):
    count = 0
    for pos, i in enumerate(y):
        if y_pred[pos] == i:
            count += 1
    acc = count / len(y)
    return acc
```

```python
accuracy(y, y_pred)
```

```python
def err_matrix(y, y_pred):
    TP = 0
    FP = 0
    FN = 0
    TN = 0
    for pos, i in enumerate(y):
        if y_pred[pos] == 1 and i == 1:
            TP += 1
        elif y_pred[pos] == 1 and i == 0:
            FP += 1
        elif y_pred[pos] == 0 and i == 1:
            FN += 1
        elif y_pred[pos] == 0 and i == 0:
            TN += 1
        else:
            print('Error')
    df = pd.DataFrame({' ':['𝑎(𝑥) = +1','𝑎(𝑥) = -1'], 'y = +1':[f'TP = {TP}', f'FN = {FN}'], 'y = -1':[f'FP = {FP}', f'TN = {TN}']})
    return df.style.hide_index()
```

```python
err_matrix(y, y_pred)
```

```python
def precision(y, y_pred):
    TP = 0
    FP = 0
    FN = 0
    TN = 0
    for pos, i in enumerate(y):
        if y_pred[pos] == 1 and i == 1:
            TP += 1
        elif y_pred[pos] == 1 and i == 0:
            FP += 1
        elif y_pred[pos] == 0 and i == 1:
            FN += 1
        elif y_pred[pos] == 0 and i == 0:
            TN += 1
        else:
            print('Error')
    return TP / (TP + FP)
```

```python
precision(y, y_pred)
```

```python
def recall(y, y_pred):
    TP = 0
    FP = 0
    FN = 0
    TN = 0
    for pos, i in enumerate(y):
        if y_pred[pos] == 1 and i == 1:
            TP += 1
        elif y_pred[pos] == 1 and i == 0:
            FP += 1
        elif y_pred[pos] == 0 and i == 1:
            FN += 1
        elif y_pred[pos] == 0 and i == 0:
            TN += 1
        else:
            print('Error')
    return TP / (TP + FN)
```

```python
recall(y, y_pred)
```
